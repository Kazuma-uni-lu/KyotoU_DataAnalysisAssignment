### Figure 1

Use `data-fig1.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig1.npy' )

plt.plot(a, marker='o',color='black')
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()
```

___

### Figure 2

Use `data-fig2.npy` for this figure.

```python
b = np.load('data-fig2.npy')
colour = ['Red', 'Green', 'Blue']
for i in range (len(b.T)):
    plt.plot(b.T[i], color=colour[i], label = colour[i])
plt.legend()
plt.xlabel("x value")
plt.ylabel("y alue")
plt.show()
```

___

### Figure 3

Use `data-fig3.npy` for this figure.

```python
c=np.load('data-fig3.npy')
plt.scatter(c[:,0],c[:,1],color = 'blue', marker='o',alpha=0.4,s=200)
plt.xlabel("x value")
plt.ylabel("y value")
plt.show()
```

___

### Figure 4

Use `data-fig4.npy` for this figure.

```python
d=np.load('data-fig4.npy')
plt.bar(x=['A','B','C','D','E'], height=d, color="gray")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

___

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
e=np.load('data-fig5.npy')
plt.imshow(e, cmap='RdBu', aspect='auto', vmin=-1, vmax=1)
plt.colorbar()
plt.show()
```

Use `data-fig-bonus-X.npy`, `data-fig-bonus-Y.npy` and `data-fig-bonus-Z.npy` for this figure.

```python
from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d
fig_x = np.load('data-fig-bonus-X.npy')
fig_y = np.load('data-fig-bonus-Y.npy')
fig_z = np.load('data-fig-bonus-Z.npy')

ax = plt.figure().add_subplot(projection='3d')
ax.contour(fig_x,fig_y,fig_z, extend3d=True, cmap=cm.coolwarm)
plt.show()
```
